```python
cur_x = 3
rate = 0.01
precision = 0.000001
previous_step_size = 1
max_iters = 10000
iters = 0
df = lambda x: 2*(x+5)

iters = 0
while (previous_step_size > precision and iters < max_iters):
    prev_x = cur_x
    cur_x = cur_x - 0.1 * df(prev_x)
    previous_step_size = abs(cur_x - prev_x)
    iters = iters + 1
    print("Iteration number {} \nCurrent x {}".format(iters, cur_x))
print("local minima occures at {}".format(cur_x))
```
